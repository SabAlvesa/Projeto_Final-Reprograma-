**Filtrando para analise, amostragem do ano 2024, Mulheres com mais de 40 anos**

# **5. Qual é a renda habitual média entre as mulheres 40+ ocupadas por estado?**



```python
df_ocupadas = df_filtrado[df_filtrado['status_mercado'] == 'Pessoas ocupadas']
df_ocupadas.groupby('uf')['renda_habitual'].agg(['mean','median','max','min'])
```

```python
estatisticas = ['mean', 'max', 'min', 'median']

renda_por_uf_raca = df_filtrado.groupby(
    ['uf', 'cor_raca']
)['renda_habitual'].agg(estatisticas).reset_index()

renda_por_uf_raca = renda_por_uf_raca.rename(columns={
    'mean': 'Renda_Media',
    'max': 'Renda_Maxima',
    'min': 'Renda_Minima',
    'median': 'Renda_Mediana'
})

print(" Estatísticas de Renda Habitual por UF e Cor/Raça")
renda_por_uf_raca.head(100)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 8))
sns.barplot(
    x='uf',
    y='Renda_Media',
    hue='cor_raca',
    data=renda_por_uf_raca,
    palette='plasma'
)

plt.title(
    'Renda Média Habitual por UF e Cor/Raça (Mulheres 40+)',
    fontsize=18,
    weight='bold'
)
plt.xlabel('Unidade da Federação (UF)', fontsize=14)
plt.ylabel('Renda Média Habitual (R$)', fontsize=14)

plt.xticks(rotation=45, ha='right')
plt.legend(title='Cor/Raça', loc='upper right')
plt.tight_layout()
plt.show()
```
